```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load the Titanic Ship Passengers Dataset
# Use the correct file path or URL for the titanic dataset.
df = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')

# 2. Check Data Types and Summary Statistics
print("Data Types:\n", df.dtypes)
print("\nSummary Statistics:\n", df.describe(include='all'))

# 3. List Column Names
print("\nColumn Names:\n", df.columns)

# 4. Exploratory Data Analysis

# Plot histogram for numeric features
df.hist(figsize=(10, 7))
plt.suptitle('Histograms of Numeric Features')
plt.show()

# Plot the proportion of survived vs. not survived
sns.countplot(x='Survived', data=df)
plt.title('Survived vs Not Survived')
plt.show()

# Plot survival rate by gender
sns.barplot(x='Sex', y='Survived', data=df)
plt.title('Survival Rate by Gender')
plt.show()

# Plot survival rate by passenger class
sns.barplot(x='Pclass', y='Survived', data=df)
plt.title('Survival Rate by Passenger Class')
plt.show()

# Plot histogram of age distribution
plt.figure(figsize=(10, 5))
sns.histplot(df['Age'].dropna(), kde=True)
plt.title('Age Distribution')
plt.show()

# Plot box plot of age by passenger class
sns.boxplot(x='Pclass', y='Age', data=df)
plt.title('Age Distribution by Passenger Class')
plt.show()

# 5. Data Wrangling

# Impute missing age values with median age
df['Age'].fillna(df['Age'].median(), inplace=True)

# Drop features such as 'Name', 'Ticket', and 'Cabin'
df = df.drop(columns=['Name', 'Ticket', 'Cabin'])

# Convert 'Sex' and 'Embarked' to numeric codes
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
df['Embarked'] = df['Embarked'].map({'C': 0, 'Q': 1, 'S': 2})

# Save the cleaned dataset for model training
df.to_csv('cleaned_titanic.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score
from sklearn.impute import SimpleImputer # Importing the SimpleImputer class for imputation

# Load the cleaned dataset
df = pd.read_csv('cleaned_titanic.csv')

# Define features and target variable
X = df[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Embarked']]
y = df['Survived']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize the imputer to replace missing values with the most frequent value
imputer = SimpleImputer(strategy='most_frequent')

# Fit and transform the imputer on the training data
X_train = imputer.fit_transform(X_train)

# Transform the test data using the trained imputer
X_test = imputer.transform(X_test)

# Initialize the logistic regression model
model = LogisticRegression(max_iter=1000)

# Fit the model to the training data
model.fit(X_train, y_train)

# Predict on the test data
y_pred = model.predict(X_test)

# Compute F1 score
f1 = f1_score(y_test, y_pred)
print(f"\nF1 Score: {f1}")

# Plot F1 Score as a Function of Penalty
penalties = ['l1', 'l2']
f1_scores = []

for penalty in penalties:
    model = LogisticRegression(penalty=penalty, solver='saga', max_iter=1000)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    f1_scores.append(f1_score(y_test, y_pred))

# Plotting
plt.scatter(penalties, f1_scores)
plt.xlabel('Penalty')
plt.ylabel('F1 Score')
plt.title('F1 Score vs Penalty')
plt.show()
```
